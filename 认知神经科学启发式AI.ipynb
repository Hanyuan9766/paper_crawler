### Unsloth

```python
from unsloth import FastLanguageModel
import torch

fourbit_models = [
    "unsloth/Qwen3-1.7B-unsloth-bnb-4bit", # Qwen 14B 2x faster
    "unsloth/Qwen3-4B-unsloth-bnb-4bit",
    "unsloth/Qwen3-8B-unsloth-bnb-4bit",
    "unsloth/Qwen3-14B-unsloth-bnb-4bit",
    "unsloth/Qwen3-32B-unsloth-bnb-4bit",

    # 4bit dynamic quants for superior accuracy and low memory use
    "unsloth/gemma-3-12b-it-unsloth-bnb-4bit",
    "unsloth/Phi-4",
    "unsloth/Llama-3.1-8B",
    "unsloth/Llama-3.2-3B",
    "unsloth/orpheus-3b-0.1-ft-unsloth-bnb-4bit" # [NEW] We support TTS models!
] # More models at https://huggingface.co/unsloth

model, tokenizer = FastLanguageModel.from_pretrained(
    model_name = "unsloth/Qwen3-14B",
    max_seq_length = 2048,   # Context length - can be longer, but uses more memory
    load_in_4bit = True,     # 4bit uses much less memory
    load_in_8bit = False,    # A bit more accurate, uses 2x memory
    full_finetuning = False, # We have full finetuning now!
    # token = "hf_...",      # use one if using gated models
)
```

We now add LoRA adapters so we only need to update 1 to 10% of all parameters!

```python
model = FastLanguageModel.get_peft_model(
    model,
    r = 32,           # Choose any number > 0! Suggested 8, 16, 32, 64, 128
    target_modules = ["q_proj", "k_proj", "v_proj", "o_proj",
                      "gate_proj", "up_proj", "down_proj",],
    lora_alpha = 32,  # Best to choose alpha = rank or rank*2
    lora_dropout = 0, # Supports any, but = 0 is optimized
    bias = "none",    # Supports any, but = "none" is optimized
    # [NEW] "unsloth" uses 30% less VRAM, fits 2x larger batch sizes!
    use_gradient_checkpointing = "unsloth", # True or "unsloth" for very long context
    random_state = 3407,
    use_rslora = False,   # We support rank stabilized LoRA
    loftq_config = None,  # And LoftQ
)
```

### Data Prep
Qwen3 has both reasoning and a non reasoning mode. So, we should use 2 datasets:

1. We use the [Open Math Reasoning]() dataset which was used to win the [AIMO](https://www.kaggle.com/competitions/ai-mathematical-olympiad-progress-prize-2/leaderboard) (AI Mathematical Olympiad - Progress Prize 2) challenge! We sample 10% of verifiable reasoning traces that used DeepSeek R1, and whicht got > 95% accuracy.

2. We also leverage [Maxime Labonne's FineTome-100k](https://huggingface.co/datasets/mlabonne/FineTome-100k) dataset in ShareGPT style. But we need to convert it to HuggingFace's normal multiturn format as well.

```python
from datasets import load_dataset
reasoning_dataset = load_dataset("unsloth/OpenMathReasoning-mini", split = "cot")
non_reasoning_dataset = load_dataset("mlabonne/FineTome-100k", split = "train")
```

Let's see the structure of both datasets:

```python
reasoning_dataset
```

```python
non_reasoning_dataset
```

We now convert the reasoning dataset into conversational format:

```python
def generate_conversation(examples):
    problems  = examples["problem"]
    solutions = examples["generated_solution"]
    conversations = []
    for problem, solution in zip(problems, solutions):
        conversations.append([
            {"role" : "user",      "content" : problem},
            {"role" : "assistant", "content" : solution},
        ])
    return { "conversations": conversations, }
```

```python
reasoning_conversations = tokenizer.apply_chat_template(
    list(reasoning_dataset.map(generate_conversation, batched = True)["conversations"]),
    tokenize = False,
)
```

Let's see the first transformed row:

```python
reasoning_conversations[0]
```

Next we take the non reasoning dataset and convert it to conversational format as well.

We have to use Unsloth's `standardize_sharegpt` function to fix up the format of the dataset first.

```python
from unsloth.chat_templates import standardize_sharegpt
dataset = standardize_sharegpt(non_reasoning_dataset)

non_reasoning_conversations = tokenizer.apply_chat_template(
    list(dataset["conversations"]),
    tokenize = False,
)
```

Let's see the first row

```python
non_reasoning_conversations[0]
```

Now let's see how long both datasets are:

```python
print(len(reasoning_conversations))
print(len(non_reasoning_conversations))
```

The non reasoning dataset is much longer. Let's assume we want the model to retain some reasoning capabilities, but we specifically want a chat model.

Let's define a ratio of chat only data. The goal is to define some mixture of both sets of data.

Let's select 75% reasoning and 25% chat based:

```python
chat_percentage = 0.25
```

Let's sample the reasoning dataset by 75% (or whatever is 100% - chat_percentage)

```python
import pandas as pd
non_reasoning_subset = pd.Series(non_reasoning_conversations)
non_reasoning_subset = non_reasoning_subset.sample(
    int(len(reasoning_conversations)*(chat_percentage/(1 - chat_percentage))),
    random_state = 2407,
)
print(len(reasoning_conversations))
print(len(non_reasoning_subset))
print(len(non_reasoning_subset) / (len(non_reasoning_subset) + len(reasoning_conversations)))
```

Finally combine both datasets:

```python
data = pd.concat([
    pd.Series(reasoning_conversations),
    pd.Series(non_reasoning_subset)
])
data.name = "text"

from datasets import Dataset
combined_dataset = Dataset.from_pandas(pd.DataFrame(data))
combined_dataset = combined_dataset.shuffle(seed = 3407)
```

<a name="Train"></a>
### Train the model
Now let's train our model. We do 60 steps to speed things up, but you can set `num_train_epochs=1` for a full run, and turn off `max_steps=None`.

```python
from trl import SFTTrainer, SFTConfig
trainer = SFTTrainer(
    model = model,
    tokenizer = tokenizer,
    train_dataset = combined_dataset,
    eval_dataset = None, # Can set up evaluation!
    args = SFTConfig(
        dataset_text_field = "text",
        per_device_train_batch_size = 2,
        gradient_accumulation_steps = 4, # Use GA to mimic batch size!
        warmup_steps = 5,
        # num_train_epochs = 1, # Set this for 1 full training run.
        max_steps = 30,
        learning_rate = 2e-4, # Reduce to 2e-5 for long training runs
        logging_steps = 1,
        optim = "adamw_8bit",
        weight_decay = 0.001,
        lr_scheduler_type = "linear",
        seed = 3407,
        report_to = "none", # Use TrackIO/WandB etc
    ),
)
```

```python
# @title Show current memory stats
gpu_stats = torch.cuda.get_device_properties(0)
start_gpu_memory = round(torch.cuda.max_memory_reserved() / 1024 / 1024 / 1024, 3)
max_memory = round(gpu_stats.total_memory / 1024 / 1024 / 1024, 3)
print(f"GPU = {gpu_stats.name}. Max memory = {max_memory} GB.")
print(f"{start_gpu_memory} GB of memory reserved.")
```

Let's train the model! To resume a training run, set `trainer.train(resume_from_checkpoint = True)`

```python
trainer_stats = trainer.train()
```

```python
# @title Show final memory and time stats
used_memory = round(torch.cuda.max_memory_reserved() / 1024 / 1024 / 1024, 3)
used_memory_for_lora = round(used_memory - start_gpu_memory, 3)
used_percentage = round(used_memory / max_memory * 100, 3)
lora_percentage = round(used_memory_for_lora / max_memory * 100, 3)
print(f"{trainer_stats.metrics['train_runtime']} seconds used for training.")
print(
    f"{round(trainer_stats.metrics['train_runtime']/60, 2)} minutes used for training."
)
print(f"Peak reserved memory = {used_memory} GB.")
print(f"Peak reserved memory for training = {used_memory_for_lora} GB.")
print(f"Peak reserved memory % of max memory = {used_percentage} %.")
print(f"Peak reserved memory for training % of max memory = {lora_percentage} %.")
```

### Inference
Let's run the model via Unsloth native inference! According to the `Qwen-3` team, the recommended settings for reasoning inference are `temperature = 0.6, top_p = 0.95, top_k = 20`

For normal chat based inference, `temperature = 0.7, top_p = 0.8, top_k = 20`

```python
messages = [
    {"role" : "user", "content" : "Solve (x + 2)^2 = 0."}
]
text = tokenizer.apply_chat_template(
    messages,
    tokenize = False,
    add_generation_prompt = True, # Must add for generation
    enable_thinking = False, # Disable thinking
)

from transformers import TextStreamer
_ = model.generate(
    **tokenizer(text, return_tensors = "pt").to("cuda"),
    max_new_tokens = 256, # Increase for longer outputs!
    temperature = 0.7, top_p = 0.8, top_k = 20, # For non thinking
    streamer = TextStreamer(tokenizer, skip_prompt = True),
)
```

```python
messages = [
    {"role" : "user", "content" : "Solve (x + 2)^2 = 0."}
]
text = tokenizer.apply_chat_template(
    messages,
    tokenize = False,
    add_generation_prompt = True, # Must add for generation
    enable_thinking = True, # Disable thinking
)

from transformers import TextStreamer
_ = model.generate(
    **tokenizer(text, return_tensors = "pt").to("cuda"),
    max_new_tokens = 1024, # Increase for longer outputs!
    temperature = 0.6, top_p = 0.95, top_k = 20, # For thinking
    streamer = TextStreamer(tokenizer, skip_prompt = True),
)
```

### GGUF / llama.cpp Conversion
To save to `GGUF` / `llama.cpp`, we support it natively now! We clone `llama.cpp` and we default save it to `q8_0`. We allow all methods like `q4_k_m`. Use `save_pretrained_gguf` for local saving and `push_to_hub_gguf` for uploading to HF.

Some supported quant methods (full list on our [Wiki page](https://github.com/unslothai/unsloth/wiki#gguf-quantization-options)):
* `q8_0` - Fast conversion. High resource use, but generally acceptable.
* `q4_k_m` - Recommended. Uses Q6_K for half of the attention.wv and feed_forward.w2 tensors, else Q4_K.
* `q5_k_m` - Recommended. Uses Q6_K for half of the attention.wv and feed_forward.w2 tensors, else Q5_K.

[**NEW**] To finetune and auto export to Ollama, try our [Ollama notebook](https://colab.research.google.com/github/unslothai/notebooks/blob/main/nb/Llama3_(8B)-Ollama.ipynb)

```python
# Save to 8bit Q8_0
if False:
    model.save_pretrained_gguf("model", tokenizer,)
# Remember to go to https://huggingface.co/settings/tokens for a token!
# And change hf to your username!
if False:
    model.push_to_hub_gguf("hf/model", tokenizer, token = "")

# Save to 16bit GGUF
if False:
    model.save_pretrained_gguf("model", tokenizer, quantization_method = "f16")
if False: # Pushing to HF Hub
    model.push_to_hub_gguf("hf/model", tokenizer, quantization_method = "f16", token = "")

# Save to q4_k_m GGUF
if False:
    model.save_pretrained_gguf("model", tokenizer, quantization_method = "q4_k_m")
if False: # Pushing to HF Hub
    model.push_to_hub_gguf("hf/model", tokenizer, quantization_method = "q4_k_m", token = "")

# Save to multiple GGUF options - much faster if you want multiple!
if False:
    model.push_to_hub_gguf(
        "hf/model", # Change hf to your username!
        tokenizer,
        quantization_method = ["q4_k_m", "q8_0", "q5_k_m",],
        token = "", # Get a token at https://huggingface.co/settings/tokens
    )
```

```python
# --- Code Block 4: Meta-Cognitive Supervisor Interface (Qwen3) ---

import re
import torch

def consult_qwen_supervisor(epoch, metrics, history_context):
    """
    ÂÖÉËÆ§Áü•Êé•Âè£ÔºöÂ∞ÜÂÆûÈ™åÊï∞ÊçÆËæìÂÖ• Qwen3ÔºåËé∑ÂèñÊéßÂà∂Êåá‰ª§„ÄÇ
    ËØ•ÂáΩÊï∞ÂÅáÂÆö 'model' Âíå 'tokenizer' Â∑≤Âú®ÂÖ®Â±Ä‰ΩúÁî®Âüü‰∏≠Âä†ËΩΩ (LLM-First Á≠ñÁï•)„ÄÇ
    """

    # 1. ÊûÑÂª∫ÁßëÂ≠¶Áä∂ÊÄÅÊä•Âëä (Scientific Status Report)
    # Â∞ÜÊäΩË±°ÁöÑÊï∞Â≠¶ÊåáÊ†áËΩ¨Âåñ‰∏∫ËÆ§Áü•Âä®ÂäõÂ≠¶ÊúØËØ≠
    status_prompt = f"""
    [ÂΩìÂâçÂÆûÈ™åÁä∂ÊÄÅ - Epoch {epoch}]
    - ‰ªªÂä°Ë°®Áé∞ (Accuracy): {metrics['acc']:.2f}% (ÁõÆÊ†á: >99%)
    - ËÆ§Áü•ÊÉäÂ•áÂ∫¶ (Avg RPE): {metrics['rpe']:.4f} (Ê≠£ÂÄº‰ª£Ë°®ÊÉäÂñúÔºåË¥üÂÄº‰ª£Ë°®Â§±Êúõ; ÁõÆÊ†á: Êî∂ÊïõËá≥ 0)
    - Êé¢Á¥¢Ê¥ªË∑ÉÂ∫¶ (Pos RPE Rate): {metrics['pos_rate']:.1f}% (ÈöèÊú∫ÁåúÊµã‰∏∫50%; Ë∂ãËøë0Ë°®Á§∫ÊéåÊè°ËßÑÂàô)
    - ËÆ§Áü•ÊµÅÂΩ¢Á®≥ÂÆöÊÄß (Max Eigenvalue |Œª|): {metrics['eigenvalue']:.4f}
      (Ê≥®: |Œª| < 1.0 Ë°®Á§∫ÂΩ¢Êàê‰∫ÜÁ®≥ÂÆöÁöÑÂê∏ÂºïÂ≠ê/Ê¶ÇÂøµ; |Œª| > 1.0 Ë°®Á§∫Ê¶ÇÂøµÂ∞öÂú®Ê≥¢Âä®)

    [ËøëÊúüÂÜ≥Á≠ñÂéÜÂè≤]
    {history_context}
    """

    # 2. ÊûÑÂª∫Á≥ªÁªüÊåá‰ª§ (System Instruction)
    # Ë¶ÅÊ±Ç Qwen3 ÊâÆÊºîÂÆûÈ™åÂØºÂ∏àÔºåÂü∫‰∫éÂä®ÂäõÂ≠¶ÁêÜËÆ∫ÂÅöÂá∫ÂÜ≥Á≠ñ
    instruction = (
        "‰Ω†ÊòØ‰∏Ä‰∏™Á≤æÈÄö BrainProp ÁÆóÊ≥ïÂíåÂä®ÂäõÁ≥ªÁªüÁêÜËÆ∫ÁöÑ AI ÂÆûÈ™åÂØºÂ∏à„ÄÇ"
        "ËØ∑Ê†πÊçÆÊèê‰æõÁöÑÊåáÊ†áÁõëÊéßÁ•ûÁªèÁΩëÁªúÁöÑ'ËÆ§Áü•ÊµÅÂΩ¢'ÂΩ¢ÊàêËøáÁ®ã„ÄÇ\n"
        "ÂÜ≥Á≠ñÈÄªËæëÂèÇËÄÉÔºö\n"
        "1. Â¶ÇÊûú |Œª| < 1.0 ‰∏î Acc > 99%ÔºåËØ¥ÊòéÊ¶ÇÂøµÂ∑≤ÂΩ¢Êàê‰∏îÁ®≥ÂÆö -> Âª∫ËÆÆ STOP„ÄÇ\n"
        "2. Â¶ÇÊûú RPE ÈúáËç°ËæÉÂ§ßÊàñ Acc ÂÅúÊªûÔºå‰ΩÜ |Œª| ‰ªçËæÉÂ§ß -> Âª∫ËÆÆ DECAY_LR (Á≤æÁªÜÊêúÁ¥¢)„ÄÇ\n"
        "3. Â¶ÇÊûúÂ§Ñ‰∫éÊó©ÊúüÂ≠¶‰π†Èò∂ÊÆµ -> Âª∫ËÆÆ CONTINUE„ÄÇ\n\n"
        "ËØ∑‰∏•Ê†ºÊåâÊ≠§Ê†ºÂºèËæìÂá∫Ôºö\n"
        "ANALYSIS: <ÁÆÄÁü≠ÂàÜÊûêÂ≠¶‰π†Áä∂ÊÄÅ>\n"
        "DECISION: <CONTINUE / DECAY_LR / STOP>\n"
        "REASON: <‰∏ÄÂè•ËØùÁêÜÁî±>"
    )

    # 3. ÊâßË°åÊé®ÁêÜ (In-Memory Inference)
    messages = [
        {"role": "system", "content": instruction},
        {"role": "user", "content": status_prompt}
    ]

    # ‰ΩøÁî® chat template Ê†ºÂºèÂåñ
    text_input = tokenizer.apply_chat_template(messages, tokenize=False, add_generation_prompt=True)
    model_inputs = tokenizer([text_input], return_tensors="pt").to(device)

    # ÁîüÊàêÂõûÂ§ç
    with torch.no_grad():
        generated_ids = model.generate(
            model_inputs.input_ids,
            max_new_tokens=256,
            temperature=0.3, # ‰ΩéÊ∏©‰ª•‰øùËØÅÊåá‰ª§ËæìÂá∫ÁöÑÁ®≥ÂÆöÊÄß
            do_sample=True
        )

    # Ëß£Á†Å
    generated_ids = [
        output_ids[len(input_ids):] for input_ids, output_ids in zip(model_inputs.input_ids, generated_ids)
    ]
    response = tokenizer.batch_decode(generated_ids, skip_special_tokens=True)[0]

    # 4. Ëß£ÊûêÁªìÊûÑÂåñÊåá‰ª§
    decision = "CONTINUE" # ÈªòËÆ§ÂÆâÂÖ®ÂõûÈÄÄ
    try:
        # ‰ºòÂÖàÂåπÈÖçÊ†áÂáÜÊ†ºÂºè
        match = re.search(r"DECISION:\s*(CONTINUE|DECAY_LR|STOP)", response, re.IGNORECASE)
        if match:
            decision = match.group(1).upper()
        else:
            # Ê®°Á≥äÂåπÈÖçÂÖúÂ∫ï
            if "STOP" in response.upper(): decision = "STOP"
            elif "DECAY" in response.upper(): decision = "DECAY_LR"

        # ÊâìÂç∞ AI ÊÄùËÄÉËøáÁ®ã (ÂÆûÊó∂ÂèçÈ¶àÁªôÊìç‰ΩúËÄÖ)
        print(f"\nü§ñ [Qwen3 Supervisor Ep{epoch}]: {decision}")
        analysis_text = response.split('DECISION')[0].replace('ANALYSIS:', '').strip()
        print(f"   Thinking: {analysis_text[:120]}...")

    except Exception as e:
        print(f"‚ö†Ô∏è Supervisor Parsing Error: {e}")

    return decision, response

print("‚úÖ Neuro-Symbolic Interface (Qwen3) Linked.")
```

```python
# --- Code Block 5: Homeostatic Perception & Cognitive Core (Self-Contained) ---

import torch
import torch.nn as nn

# [ÂÖ≥ÈîÆ‰øÆÊ≠£] ÈáçÊñ∞ÂÆö‰πâËÆæÂ§áÔºåÈò≤Ê≠¢Âõ†ÈáçÂêØËøêË°åÊó∂ÂØºËá¥ÂèòÈáè‰∏¢Â§±
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Device re-initialized: {device}")

class BioLinear(nn.Linear):
    """
    ÁîüÁâ©Á∫øÊÄßÂ±Ç (Biologically Plausible Linear Layer).
    Ê†∏ÂøÉÁâπÊÄßÔºöÂú®ÂâçÂêë‰º†Êí≠‰∏≠'ËÆ∞ÂøÜ'Á™ÅËß¶ÂâçÊ¥ªÊÄß (Presynaptic Activity).
    ËøôÊòØ BrainProp ‰∏âÂõ†Á¥†Êõ¥Êñ∞ËßÑÂàôÁöÑÂøÖË¶ÅÁªÑ‰ª∂: Œîw = RPE * Feedback * Activity.
    """
    def __init__(self, in_features, out_features, bias=True):
        super(BioLinear, self).__init__(in_features, out_features, bias)
        self.input_activity = None

    def forward(self, input):
        # 1. ËÆ∞ÂΩïÁ™ÅËß¶ÂâçÊ¥ªÊÄß (detach Èò≤Ê≠¢Âπ≤Êâ∞Ëá™Âä®ÂæÆÂàÜÂõæÔºå‰ªÖÁî®‰∫éËµ´Â∏ÉËßÑÂàôËÆ°ÁÆó)
        self.input_activity = input.detach()
        # 2. Ê†áÂáÜÁ∫øÊÄßÂèòÊç¢
        return super(BioLinear, self).forward(input)

class PerceptionNetwork(nn.Module):
    """
    Á®≥ÊÄÅÊÑüÁü•ÁΩëÁªú (Homeostatic Perception Network).
    ÁªìÊûÑ: Linear -> BatchNorm -> ReLU.
    ÂºïÂÖ• BatchNorm ÊòØ‰∏∫‰∫ÜÊ®°ÊãüÁ®≥ÊÄÅÂèØÂ°ëÊÄßÔºåÈò≤Ê≠¢ÂèçÈ¶à‰ø°Âè∑Âú®Ê∑±Â±ÇË°∞Âáè.
    """
    def __init__(self, input_dim, hidden_dims=[256, 128]):
        super(PerceptionNetwork, self).__init__()
        self.layers = nn.ModuleList()

        current_dim = input_dim
        for h_dim in hidden_dims:
            self.layers.append(nn.ModuleDict({
                'linear': BioLinear(current_dim, h_dim),
                'bn': nn.BatchNorm1d(h_dim),  # [ÂÖ≥ÈîÆÊú∫Âà∂] Á®≥ÊÄÅË∞ÉËäÇ
                'act': nn.ReLU()              # Èó®ÊéßÊú∫Âà∂
            }))
            current_dim = h_dim

        self.output_dim = current_dim

    def forward(self, x):
        # Â±ïÂπ≥ËæìÂÖ• (ÈÄÇÈÖç MNIST)
        if x.dim() > 2:
            x = x.view(x.size(0), -1)

        for layer in self.layers:
            x = layer['linear'](x)
            x = layer['bn'](x)   # ÂΩí‰∏ÄÂåñÊ¥ªÊÄß
            x = layer['act'](x)  # Á®ÄÁñèÊøÄÊ¥ª

        return x

class CognitiveBottleneck(nn.Module):
    """
    ËÆ§Áü•Âä®ÂäõÂ≠¶Áì∂È¢à (TinyRNN Core).
    ÂäüËÉΩ:
    1. Êó∂Â∫èÁßØÂàÜ: Â∞ÜÊÑüÁü•ÁâπÂæÅËΩ¨Âåñ‰∏∫ËÆ§Áü•Áä∂ÊÄÅ (h_t).
    2. Âº∫Âà∂ÈôçÁª¥: Âú®‰∫åÁª¥ÊµÅÂΩ¢‰∏äÊûÑÂª∫Âê∏ÂºïÂ≠ê (Attractors).
    """
    def __init__(self, input_size, hidden_size):
        super(CognitiveBottleneck, self).__init__()
        self.hidden_size = hidden_size

        # ‰ΩøÁî® GRUCell ‰ª•‰æøÁ≤æÁªÜÊéßÂà∂ÂçïÊ≠•Âä®ÂäõÂ≠¶ÂíåÁä∂ÊÄÅÊö¥Èú≤
        self.rnn_cell = nn.GRUCell(input_size, hidden_size)

        # ËÆ∞ÂΩïÂ±ÄÈÉ®Ê¥ªÊÄßÁî®‰∫éÊõ¥Êñ∞
        self.input_activity = None
        self.prev_hidden = None

        # Âä®ÂäõÂ≠¶ËΩ®ËøπËÆ∞ÂΩïÂô® (Áî®‰∫éÁªòÂà∂Áõ∏Âõæ)
        self.state_history = []

    def forward(self, input_feature, h_prev):
        # 1. ËÆ∞ÂΩïÁ™ÅËß¶ÂâçÊ¥ªÊÄß (ÁâπÂæÅËæìÂÖ• & ‰∏ä‰∏ÄÊó∂ÂàªÁä∂ÊÄÅ)
        self.input_activity = input_feature.detach()
        self.prev_hidden = h_prev.detach()

        # 2. ÊâßË°åÂä®ÂäõÂ≠¶Êõ¥Êñ∞: h_t = GRU(x_t, h_{t-1})
        h_curr = self.rnn_cell(input_feature, h_prev)

        # 3. ‰ªÖÂú®ÈùûËÆ≠ÁªÉÊ®°Âºè‰∏ãËÆ∞ÂΩïËΩ®Ëøπ (ÈÅøÂÖçÊòæÂ≠òÊ∫¢Âá∫)
        if not self.training:
             self.state_history.append(h_curr.detach().cpu().numpy())

        return h_curr

    def reset_state_history(self):
        """Ê∏ÖÁ©∫ÂéÜÂè≤ËÆ∞ÂΩïÔºå‰∏∫Êñ∞‰∏ÄËΩÆÂä®ÂäõÂ≠¶ÂàÜÊûêÂÅöÂáÜÂ§á"""
        self.state_history = []

# ÂÆû‰æãÂåñÊ®°ÂûãÂπ∂ËΩ¨ÁßªÂà∞ A100 GPU
# Ê≥®ÊÑèÔºöcfg ÂèòÈáèÂ∫îËØ•Âú® Block 3 ‰∏≠ÂÆö‰πâËøáÔºåÂ¶ÇÊûúÊä•Èîô cfg Êú™ÂÆö‰πâÔºåËØ∑ÈáçÊñ∞ËøêË°å Block 3
perception_net = PerceptionNetwork(input_dim=cfg.input_dim).to(device)
cognitive_net = CognitiveBottleneck(
    input_size=perception_net.output_dim,
    hidden_size=cfg.hidden_size # d=2
).to(device)

print(f"Subject Architecture Built.\n- Perception: BatchNorm Enabled\n- Cognition: TinyRNN (d={cfg.hidden_size})")
```

```python
# --- Code Block 6: Action Readout & Stochastic Policy ---

class ActionReadout(nn.Module):
    """
    Âä®‰ΩúËØªÂá∫Â±Ç (Action Readout Layer).
    ÂäüËÉΩ:
      1. Â∞ÜËÆ§Áü•Áä∂ÊÄÅ (h_t) Êò†Â∞Ñ‰∏∫Âä®‰ΩúÂæóÂàÜ (Logits).
      2. ÊâßË°åÈöèÊú∫ÈááÊ†∑ (Stochastic Sampling) ‰ª•Ê®°ÊãüÁîüÁâ©Êé¢Á¥¢Ë°å‰∏∫.
      3. ËæìÂá∫Ë¢´ÈÄâÂä®‰ΩúÁöÑÊ¥ªÊÄßÂÄº‰Ωú‰∏∫'È¢ÑÊúüÂ•ñËµè' (Expected Reward).
    """
    def __init__(self, hidden_size, num_actions):
        super(ActionReadout, self).__init__()
        # ‰ΩøÁî® BioLinearÔºåÂõ†‰∏∫Ëøô‰∏ÄÂ±ÇÁöÑÊùÉÈáç (W_out) ‰πüË¶ÅÈÄöËøá‰∏âÂõ†Á¥†ËßÑÂàôÊõ¥Êñ∞
        # Êõ¥Êñ∞ÂÖ¨Âºè: ŒîW_out = RPE * 1 * h_t (ÂÖ∂‰∏≠ Feedback=1, Activity=h_t)
        self.linear = BioLinear(hidden_size, num_actions)
        self.softmax = nn.Softmax(dim=1)

    def forward(self, h_curr):
        # 1. ËÆ°ÁÆóÂä®‰ΩúÂæóÂàÜ (Scores/Logits)
        scores = self.linear(h_curr)

        # 2. ËÆ°ÁÆóÂä®‰ΩúÊ¶ÇÁéá (Probabilities)
        # Âú® BrainProp ‰∏≠ÔºåËøô‰∏™Ê¶ÇÁéáÂÄº y_s Ë¢´ËßÜ‰∏∫ÂØπÂ•ñËµèÁöÑÈ¢ÑÊúü (Value Estimate)
        probs = self.softmax(scores)
        return probs, scores

    def select_action(self, probs):
        """
        ÊâßË°åÈöèÊú∫Âä®‰ΩúÈááÊ†∑ (Stochastic Sampling).
        BrainProp ‰æùËµñËøôÁßçËØïÈîôÊú∫Âà∂ (Trial-and-Error) Êù•Êé¢Á¥¢ÁéØÂ¢É.
        """
        # ÂàõÂª∫ÂàÜÁ±ªÂàÜÂ∏É (Categorical Distribution)
        dist = torch.distributions.Categorical(probs)

        # ÈááÊ†∑Âä®‰ΩúÁ¥¢Âºï (Batch_Size,)
        action = dist.sample()

        # ÊèêÂèñË¢´ÈÄâÂä®‰ΩúÁöÑÊ¶ÇÁéá (Selected Probability)
        # Áî®‰∫éÂêéÁª≠ËÆ°ÁÆó RPE: Œ¥ = r - p_selected
        # gather Êìç‰ΩúÈÄâÂèñÊØè‰∏™Ê†∑Êú¨ÂÆûÈôÖÊâßË°åÂä®‰ΩúÂØπÂ∫îÁöÑÊ¶ÇÁéáÂÄº
        selected_prob = probs.gather(1, action.view(-1, 1))

        return action, selected_prob

# ÂÆû‰æãÂåñËØªÂá∫Â±Ç
# hidden_size: ÂåπÈÖç CognitiveBottleneck ÁöÑËæìÂá∫ (d=2)
# num_actions: ÂåπÈÖçËØæÁ®ãÂ≠¶‰π†‰ªªÂä°ÁöÑÁ±ªÂà´Êï∞ (2ÂàÜÁ±ª)
# Ê≥®ÊÑè: cfg ÂèòÈáèÊù•Ëá™ Block 3
readout_net = ActionReadout(
    hidden_size=cfg.hidden_size,
    num_actions=cfg.num_classes
).to(device)

print(f"Readout Layer initialized (Output Actions={cfg.num_classes}).")
print("  - Mechanism: Stochastic Sampling (Trial-and-Error)")
```

```python
# --- Code Block 7: High-Velocity Curriculum Data Stream ---

from torchvision import datasets, transforms
from torch.utils.data import Dataset, DataLoader

class CurriculumMNIST(Dataset):
    """
    ËØæÁ®ãÂ≠¶‰π†‰∏ìÁî®Êï∞ÊçÆÈõÜ.
    ÂäüËÉΩÔºöËøáÊª§ÁâπÂÆöÊï∞Â≠óÂπ∂ÈáçÊò†Â∞ÑÊ†áÁ≠æÔºåÊûÑÂª∫ÁÆÄÂåñÁöÑËÆ§Áü•‰ªªÂä°ÁéØÂ¢É.
    """
    def __init__(self, root, train=True, target_digits=[0, 1], transform=None):
        # ‰∏ãËΩΩÂéüÂßã MNIST
        self.base_dataset = datasets.MNIST(root, train=train, download=True, transform=transform)
        self.target_digits = sorted(target_digits)

        # Âª∫Á´ãÊò†Â∞ÑË°® (e.g., {0:0, 1:1} or {3:0, 7:1})
        # Ëøô‰∏ÄÊ≠•Ëá≥ÂÖ≥ÈáçË¶ÅÔºå‰øùËØÅ‰∫Ü output layer ÁöÑÁ¨¨ 0 ‰∏™Á•ûÁªèÂÖÉÂØπÂ∫î target_digits[0]
        self.label_map = {digit: i for i, digit in enumerate(self.target_digits)}

        # È¢ÑÁ≠õÈÄâÁ¥¢Âºï
        self.indices = [i for i, target in enumerate(self.base_dataset.targets)
                        if int(target) in self.target_digits]

        split = "Train" if train else "Test"
        print(f"[{split}] Filtered digits {target_digits}. Samples: {len(self.indices)}")

    def __len__(self):
        return len(self.indices)

    def __getitem__(self, idx):
        original_idx = self.indices[idx]
        img, target = self.base_dataset[original_idx]

        # ÂÖ≥ÈîÆÔºöÈáçÊò†Â∞ÑÊ†áÁ≠æ‰ª•ÈÄÇÈÖç Readout Layer Áª¥Â∫¶
        return img, self.label_map[int(target)]

def get_curriculum_loaders():
    # Ê†áÂáÜÂåñÂØπ BrainProp ÁöÑ ReLU Èó®ÊéßÊú∫Âà∂Ëá≥ÂÖ≥ÈáçË¶Å
    transform = transforms.Compose([
        transforms.ToTensor(),
        transforms.Normalize((0.1307,), (0.3081,))
    ])

    train_set = CurriculumMNIST('./data', train=True, target_digits=cfg.target_digits, transform=transform)
    test_set = CurriculumMNIST('./data', train=False, target_digits=cfg.target_digits, transform=transform)

    # [A100 ‰ºòÂåñ] ÂºÄÂêØÂ§öÁ∫øÁ®ãÈ¢ÑÂèñ‰∏éÈîÅÈ°µÂÜÖÂ≠ò
    return (DataLoader(train_set, batch_size=cfg.batch_size, shuffle=True,
                       num_workers=cfg.num_workers, pin_memory=True),
            DataLoader(test_set, batch_size=cfg.batch_size, shuffle=False,
                       num_workers=cfg.num_workers, pin_memory=True))

# ÂÆû‰æãÂåñÊï∞ÊçÆÊµÅ
train_loader, test_loader = get_curriculum_loaders()
print("üöÄ High-Velocity Data Stream Ready.")
```

```python
# --- Code Block 8: Dynamic Optimizer & Dynamical Diagnostics ---

import torch.autograd.functional as F_grad
from scipy.optimize import root

class BrainPropOptimizer:
    """
    ÊîØÊåÅ AI Âπ≤È¢ÑÁöÑÂä®ÊÄÅ BrainProp ‰ºòÂåñÂô®.
    """
    def __init__(self, parameters, learning_rate):
        self.parameters = list(parameters)
        self.lr = learning_rate

    def step(self, rpe):
        """
        ÊâßË°å‰∏ÄÊ≠•ÊùÉÈáçÊõ¥Êñ∞.
        ÂéüÁêÜ: Œîw = Œ∑ * RPE * (Feedback * Activity)
        PyTorch ÁöÑ .grad Â∑≤ÁªèÂåÖÂê´‰∫Ü Feedback * Activity (ÈÄöËøá‰º™ÊçüÂ§± backward ÂæóÂà∞)
        """
        rpe_val = rpe.item() if isinstance(rpe, torch.Tensor) else rpe

        with torch.no_grad():
            for p in self.parameters:
                if p.grad is not None:
                    # Ê¢ØÂ∫¶‰∏äÂçá (Maximize Reward)
                    p.data += self.lr * rpe_val * p.grad

        # Ê∏ÖÈô§Ê¢ØÂ∫¶Ôºå‰∏∫‰∏ã‰∏ÄËΩÆ Feedback ‰º†Êí≠ÂÅöÂáÜÂ§á
        self.zero_grad()

    def zero_grad(self):
        for p in self.parameters:
            if p.grad is not None:
                p.grad.detach_()
                p.grad.zero_()

    def set_learning_rate(self, new_lr):
        """AI Âπ≤È¢ÑÊé•Âè£ÔºöÂÖÅËÆ∏ Qwen3 Âä®ÊÄÅË∞ÉÊï¥Â≠¶‰π†Áéá"""
        print(f"‚ö° [Optimizer] Learning Rate adjusted: {self.lr:.4f} -> {new_lr:.4f}")
        self.lr = new_lr

def get_attractor_stability(perception_net, cognitive_net, loader, target_idx=0):
    """
    Âä®ÂäõÂ≠¶ËØäÊñ≠Âô®ÔºöËÆ°ÁÆóËÆ§Áü•ÊµÅÂΩ¢ÁöÑ‰∏çÂä®ÁÇπÁ®≥ÂÆöÊÄß (|Œª|).
    ËøôÊòØ Qwen3 Âà§Êñ≠'Ê¶ÇÂøµÊòØÂê¶ÂΩ¢Êàê'ÁöÑÊ†∏ÂøÉÁâ©ÁêÜ‰æùÊçÆ.
    """
    cognitive_net.eval(); perception_net.eval()

    # 1. ÊèêÂèñÂéüÂûãÁâπÂæÅ (Prototype)
    # ËÆ°ÁÆóÁõÆÊ†áÊï∞Â≠óÂú®ÊÑüÁü•Â±ÇËæìÂá∫ÁöÑÂπ≥ÂùáÁâπÂæÅÂêëÈáè
    proto = None
    with torch.no_grad():
        for imgs, lbls in loader:
            imgs = imgs.to(device)
            mask = (lbls == target_idx)
            if mask.sum() > 0:
                proto = perception_net(imgs[mask]).mean(dim=0).unsqueeze(0)
                break

    if proto is None: return 1.5 # Êó†Êï∞ÊçÆÔºåËßÜ‰∏∫‰∏çÁ®≥ÂÆö

    # 2. ÂØªÊâæ‰∏çÂä®ÁÇπ (Fixed Point Search)
    # Ê±ÇËß£ÊñπÁ®ã: F(h) = RNN(h, x) - h = 0
    def dynamics(h_np):
        h_t = torch.tensor(h_np, dtype=torch.float32).unsqueeze(0).to(device)
        with torch.no_grad():
            h_next = cognitive_net(proto, h_t)
        return (h_next - h_t).cpu().numpy().flatten()

    # ‰ΩøÁî® Hybr ÊñπÊ≥ïÊ±ÇËß£ÈùûÁ∫øÊÄßÊñπÁ®ãÁªÑ
    sol = root(dynamics, np.zeros(cfg.hidden_size), method='hybr')

    if not sol.success: return 1.5 # Ê±ÇËß£Â§±Ë¥•

    # 3. ËÆ°ÁÆóÈõÖÂèØÊØîÁü©ÈòµÁâπÂæÅÂÄº (Stability Analysis)
    fp = torch.tensor(sol.x, dtype=torch.float32).unsqueeze(0).to(device)

    # ÂÆö‰πâÂçïÊ≠•Âä®ÂäõÂ≠¶ÂáΩÊï∞Áî®‰∫éÊ±ÇÂØº
    def step_func(h):
        return cognitive_net(proto, h)

    try:
        # Ëá™Âä®ÂæÆÂàÜËÆ°ÁÆó Jacobian
        jac = F_grad.jacobian(step_func, fp).squeeze()

        # ËÆ°ÁÆóÁâπÂæÅÂÄºÊ®°ÁöÑÊúÄÂ§ßÂÄº
        if jac.dim() == 0:
            eig = jac.abs().item()
        else:
            eig = torch.linalg.eigvals(jac).abs().max().item()

        return eig
    except:
        return 1.5

# ÂàùÂßãÂåñ‰ºòÂåñÂô®
# Êî∂ÈõÜÊâÄÊúâÊ®°ÂùóÂèÇÊï∞ (ÊÑüÁü•„ÄÅËÆ§Áü•„ÄÅËØªÂá∫)
all_params = (
    list(perception_net.parameters()) +
    list(cognitive_net.parameters()) +
    list(readout_net.parameters())
)
optimizer = BrainPropOptimizer(all_params, cfg.learning_rate)

print("‚úÖ Optimizer & Dynamical Diagnostics Ready.")
```

```python
# --- Code Block 9: AI-Supervised Training Loop (Fixed Imports) ---

import time
import torch
import numpy as np

# 1. ËæÖÂä©ÂáΩÊï∞ÔºöRPE ËÆ°ÁÆó
def calc_rpe(selected_probs, actions, targets):
    """
    ËÆ°ÁÆó BrainProp Ê†∏ÂøÉ‰ø°Âè∑:
    Reward: ÁåúÂØπ=1, ÁåúÈîô=0
    RPE: Reward - Expected_Value (Âç≥ Selected_Prob)
    """
    rewards = (actions == targets).float()
    # flattenÂ§ÑÁêÜ‰ª•Èò≤Áª¥Â∫¶‰∏çÂåπÈÖç
    expectations = selected_probs.view(-1)
    rpes = rewards - expectations
    return rewards, rpes

# 2. ÂàùÂßãÂåñÊó•ÂøóÂÆπÂô®
logs = {'acc': [], 'rpe': [], 'pos_rate': [], 'eigenvalue': []}
dynamics_history = {}
supervisor_log = [] # ËÆ∞ÂΩï Qwen3 ÁöÑÊÄùÁª¥Èìæ

print(f"ü§ñ AI-Guided Experiment Started.\n   Supervisor: Qwen3-14B\n   Task: Binary Classification {cfg.target_digits}")

start_time = time.time()

for epoch in range(cfg.num_epochs):
    # --- [Part A] BrainProp Micro-Loop (ÁîüÁâ©Â≠¶‰π†) ---
    perception_net.train(); cognitive_net.train(); readout_net.train()
    stats = {'corr': 0, 'rpe': 0, 'pos': 0, 'n': 0}

    for x, y in train_loader:
        x, y = x.to(device), y.to(device)
        bs = x.size(0)

        # 1. ÂâçÂêë‰º†Êí≠ (Forward)
        # h_0 ÂàùÂßãÂåñ‰∏∫ 0
        h_curr = torch.zeros(bs, cfg.hidden_size).to(device)
        features = perception_net(x)
        h_curr = cognitive_net(features, h_curr) # ÈÄöËøá d=2 Áì∂È¢à
        probs, scores = readout_net(h_curr)

        # 2. Âä®‰ΩúÈááÊ†∑‰∏éÂèçÈ¶à (Action & Feedback)
        act, p_sel = readout_net.select_action(probs)
        rewards, rpes = calc_rpe(p_sel, act, y)

        # 3. Ê≥®ÊÑèÂäõÂèçÈ¶à‰º†Êí≠ (Attention Backward)
        # ‰ªÖÂØπ"Ë¢´ÈÄâÂä®‰Ωú"ÁöÑÂæóÂàÜÊ±ÇÂØºÔºåÊ®°ÊãüÁîüÁâ©Ê≥®ÊÑèÂäõÂõû‰º†
        optimizer.zero_grad()
        selected_scores = scores.gather(1, act.view(-1, 1))
        selected_scores.sum().backward()

        # 4. ÊùÉÈáçÊõ¥Êñ∞ (Synaptic Update)
        # Œîw = lr * RPE * Gradient(Feedback)
        optimizer.step(rpes.mean())

        # ÁªüËÆ°
        stats['corr'] += (act == y).sum().item()
        stats['rpe'] += rpes.mean().item()
        stats['pos'] += (rpes > 0).sum().item() # Ê≠£ÂêëÂèçÈ¶àËÆ°Êï∞
        stats['n'] += bs

    # ËÆ°ÁÆóÊú¨ËΩÆÊåáÊ†á
    ep_acc = 100 * stats['corr'] / stats['n']
    ep_rpe = stats['rpe'] / len(train_loader)
    ep_pos = 100 * stats['pos'] / stats['n']

    # --- [Part B] AI Supervision Interval (ÂÖÉËÆ§Áü•Âπ≤È¢Ñ) ---
    # ÊØè N ËΩÆÊàñÊúÄÂêé‰∏ÄËΩÆËøõË°å‰∏ÄÊ¨°ËØäÊñ≠
    if epoch % cfg.consult_interval == 0 or epoch == cfg.num_epochs - 1:
        print(f"\nüîç [Epoch {epoch}] Pausing for AI Diagnostics...")

        # 1. Áâ©ÁêÜËØäÊñ≠ÔºöËÆ°ÁÆóÂê∏ÂºïÂ≠êÁ®≥ÂÆöÊÄß
        eig_val = get_attractor_stability(perception_net, cognitive_net, test_loader)

        # 2. Âí®ËØ¢ AI ÂØºÂ∏à
        metrics_snap = {'acc': ep_acc, 'rpe': ep_rpe, 'pos_rate': ep_pos, 'eigenvalue': eig_val}
        # ‰º†ÂÖ•ÊúÄËøë 3 Êù°ÂÜ≥Á≠ñÂéÜÂè≤‰Ωú‰∏∫‰∏ä‰∏ãÊñá
        history_ctx = "\n".join(supervisor_log[-3:]) if supervisor_log else "None"
        decision, reasoning = consult_qwen_supervisor(epoch, metrics_snap, history_ctx)

        # 3. ÊâßË°åÂπ≤È¢Ñ
        if decision == "STOP":
            print(f"üõë [Intervention] Qwen3 triggered EARLY STOPPING. (Reason: {reasoning[:50]}...)")
            logs['eigenvalue'].append(eig_val) # ËÆ∞ÂΩïÊúÄÂêé‰∏ÄÊ¨°ÁâπÂæÅÂÄº
            break # Ë∑≥Âá∫ËÆ≠ÁªÉÂæ™ÁéØ

        elif decision == "DECAY_LR":
            new_lr = optimizer.lr * 0.5
            optimizer.set_learning_rate(new_lr)

        # 4. ËÆ∞ÂΩïÊó•Âøó‰∏éÂø´ÁÖß
        supervisor_log.append(f"Ep{epoch}: {decision} | Œª={eig_val:.2f} | {reasoning[:50]}...")
        logs['eigenvalue'].append(eig_val)

        # ËÆ∞ÂΩïÂä®ÂäõÂ≠¶Áä∂ÊÄÅÁî®‰∫éÁªòÂõæ (Snapshot)
        cognitive_net.eval(); cognitive_net.reset_state_history()
        # Âèñ‰∏ÄÊâπÊµãËØïÊï∞ÊçÆË∑ë‰∏ÄÈÅçÂâçÂêëÔºåËÆ∞ÂΩïËΩ®Ëøπ
        with torch.no_grad():
            tx, ty = next(iter(test_loader))
            tx = tx.to(device)
            tf = perception_net(tx)
            th = cognitive_net(tf, torch.zeros(tx.size(0), cfg.hidden_size).to(device))
        # ‰øùÂ≠ò (Hidden_States, Labels)
        dynamics_history[epoch] = (np.concatenate(cognitive_net.state_history), ty.numpy())

    else:
        # ÈùûÂí®ËØ¢ËΩÆÊ¨°ÔºåÁâπÂæÅÂÄºÂ°´Á©∫(ÊàñÊ≤øÁî®‰∏ä‰∏ÄÊ¨°)‰ª•‰øùÊåÅÊï∞ÁªÑÈïøÂ∫¶ÂØπÈΩê
        pass

    # ËÆ∞ÂΩïÂ∏∏ËßÑÊó•Âøó
    logs['acc'].append(ep_acc); logs['rpe'].append(ep_rpe); logs['pos_rate'].append(ep_pos)
    print(f"   Ep {epoch}: Acc {ep_acc:.1f}% | RPE {ep_rpe:.3f} | PosRate {ep_pos:.1f}%")

print(f"‚úÖ Training Session Ended in {time.time()-start_time:.1f}s")
```

```python
# --- Code Block 10 (Fixed): Comprehensive Visualization Suite with Path Recovery ---

import matplotlib.pyplot as plt
import matplotlib.animation as animation
import matplotlib.gridspec as gridspec
import seaborn as sns
import numpy as np
import torch
import os
from datetime import datetime

# --- [ÂÖ≥ÈîÆ‰øÆÂ§ç] Ë∑ØÂæÑÂèòÈáèËá™ÊÑàÊú∫Âà∂ ---
# Èò≤Ê≠¢Âõ†ÈáçÂêØËøêË°åÊó∂ÂØºËá¥ plot_dir ‰∏¢Â§±Êä•Èîô
if 'plot_dir' not in globals():
    print("‚ö†Ô∏è Warning: 'plot_dir' variable lost. Attempting to recover...")

    # Â∞ùËØïÊü•ÊâæÊàñÂàõÂª∫ÈªòËÆ§Ë∑ØÂæÑ
    base_path = '/content/drive/MyDrive/AG_DBN_Experiments'

    # Â¶ÇÊûú experiment_dir ËøòÂú®ÔºåÂ∞±Áî®ÂÆÉ
    if 'experiment_dir' in globals():
        plot_dir = os.path.join(experiment_dir, 'plots')
    else:
        # Â¶ÇÊûúÈÉΩ‰∏çÂú®‰∫ÜÔºåÂàõÂª∫‰∏Ä‰∏™Êñ∞ÁöÑÊÅ¢Â§çÁõÆÂΩï
        timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
        recovery_dir = os.path.join(base_path, f"Exp_Restored_{timestamp}")
        plot_dir = os.path.join(recovery_dir, 'plots')
        # ÂêåÊó∂ÊÅ¢Â§çÂÖ∂‰ªñË∑ØÂæÑÂèòÈáè‰ª•Èò≤‰∏á‰∏Ä
        experiment_dir = recovery_dir

    if not os.path.exists(plot_dir):
        os.makedirs(plot_dir, exist_ok=True)
    print(f"‚úÖ Path recovered. Saving plots to: {plot_dir}")

# ËÆæÁΩÆÁªòÂõæÈ£éÊ†º
plt.style.use('seaborn-v0_8-darkgrid')

def get_prototypes(perception_net, loader, target_indices):
    """ËæÖÂä©ÂáΩÊï∞ÔºöËÆ°ÁÆóÁõÆÊ†áÁ±ªÂà´ÁöÑÂéüÂûãÁâπÂæÅ"""
    prototypes = {}
    perception_net.eval()
    with torch.no_grad():
        # Â∞ùËØïËé∑ÂèñÊï∞ÊçÆËø≠‰ª£Âô®
        try:
            for imgs, lbls in loader:
                imgs = imgs.to(device)
                feats = perception_net(imgs)
                for idx in target_indices:
                    mask = (lbls == idx)
                    if mask.sum() > 0:
                        if idx not in prototypes: prototypes[idx] = []
                        prototypes[idx].append(feats[mask])
                if len(prototypes) == len(target_indices) and all(len(p) > 5 for p in prototypes.values()):
                    break
        except Exception as e:
            print(f"‚ö†Ô∏è Prototype extraction warning: {e}")
            return {}

    final_protos = {}
    for idx, feat_list in prototypes.items():
        if len(feat_list) > 0:
            final_protos[idx] = torch.cat(feat_list, dim=0).mean(dim=0).unsqueeze(0)
    return final_protos

def compute_vector_field_and_heatmap(cognitive_net, readout_net, prototype, resolution=30, limit=1.8):
    """ËæÖÂä©ÂáΩÊï∞ÔºöËÆ°ÁÆóÊµÅÂú∫(Vector Field)ÂíåÁΩÆ‰ø°Â∫¶ÁÉ≠Âäõ(Heatmap)"""
    x_lin = np.linspace(-limit, limit, resolution)
    y_lin = np.linspace(-limit, limit, resolution)
    X, Y = np.meshgrid(x_lin, y_lin)

    grid_tensor = torch.zeros(resolution*resolution, cfg.hidden_size).to(device)
    for i in range(resolution):
        for j in range(resolution):
            grid_tensor[i*resolution+j] = torch.tensor([X[i,j], Y[i,j]], dtype=torch.float32)

    with torch.no_grad():
        # TinyRNN Âä®ÂäõÂ≠¶
        h_next = cognitive_net(prototype.repeat(resolution*resolution, 1), grid_tensor)
        delta = (h_next - grid_tensor).cpu().numpy()
        U = delta[:, 0].reshape(resolution, resolution)
        V = delta[:, 1].reshape(resolution, resolution)

        # BrainProp ÂÖÉËÆ§Áü• (ÁΩÆ‰ø°Â∫¶)
        probs, _ = readout_net(grid_tensor)
        confidences = probs.max(dim=1)[0].view(resolution, resolution).cpu().numpy()

    return X, Y, U, V, confidences

# --- ÂäüËÉΩ 1: ÈùôÊÄÅÂø´ÁÖß ---
def save_static_snapshots(dynamics_history, perception_net, cognitive_net, readout_net, save_dir):
    print("üì∏ Saving static phase portraits...")
    target_indices = sorted(cfg.target_digits)
    prototypes = get_prototypes(perception_net, test_loader, range(len(target_indices)))

    if not prototypes:
        print("‚ö†Ô∏è No prototypes found. Skipping static plots.")
        return

    snapshot_dir = os.path.join(save_dir, 'snapshots_hd')
    os.makedirs(snapshot_dir, exist_ok=True)

    sorted_epochs = sorted(dynamics_history.keys())
    for epoch in sorted_epochs:
        h_real, y_real = dynamics_history[epoch]
        fig, axes = plt.subplots(1, 2, figsize=(14, 6))

        cognitive_net.eval(); readout_net.eval()
        for i, idx in enumerate(range(len(target_indices))):
            if i >= len(axes): break
            ax = axes[i]
            if idx not in prototypes: continue

            proto = prototypes[idx]
            X, Y, U, V, conf = compute_vector_field_and_heatmap(cognitive_net, readout_net, proto)

            ax.imshow(conf, extent=[-1.8, 1.8, -1.8, 1.8], origin='lower', cmap='Greys', alpha=0.3, vmin=0.5, vmax=1.0)
            speed = np.sqrt(U**2 + V**2)
            ax.streamplot(X, Y, U, V, color=speed, cmap='coolwarm', density=1.2, linewidth=1)

            if (y_real == idx).sum() > 0:
                h_cls = h_real[y_real == idx]
                ax.scatter(h_cls[:100,0], h_cls[:100,1], c='black', s=15, edgecolors='white', alpha=0.7)

            digit = cfg.target_digits[idx]
            ax.set_title(f"Digit {digit} Attractor (Epoch {epoch})")
            ax.set_xlim(-1.8, 1.8); ax.set_ylim(-1.8, 1.8)

        plt.suptitle(f"Cognitive Manifold Snapshot - Epoch {epoch}")
        plt.tight_layout()
        plt.savefig(os.path.join(snapshot_dir, f"manifold_epoch_{epoch:03d}.png"), dpi=100)
        plt.close()
    print(f"‚úÖ Static snapshots saved to: {snapshot_dir}")

# --- ÂäüËÉΩ 2: Âä®ÊÄÅ GIF ---
def generate_evolution_gif(dynamics_history, perception_net, cognitive_net, readout_net, save_dir):
    print("üé• Rendering evolution GIF...")
    target_indices = sorted(cfg.target_digits)
    prototypes = get_prototypes(perception_net, test_loader, range(len(target_indices)))
    if not prototypes: return

    sorted_epochs = sorted(dynamics_history.keys())
    fig, axes = plt.subplots(1, 2, figsize=(12, 5))

    def update(frame_epoch):
        h_real, y_real = dynamics_history[frame_epoch]
        for ax in axes: ax.clear()

        for i, idx in enumerate(range(len(target_indices))):
            if i >= len(axes): break
            ax = axes[i]
            if idx not in prototypes: continue

            proto = prototypes[idx]
            X, Y, U, V, conf = compute_vector_field_and_heatmap(cognitive_net, readout_net, proto)

            ax.imshow(conf, extent=[-1.8, 1.8, -1.8, 1.8], origin='lower', cmap='magma', alpha=0.4, vmin=0.5, vmax=1.0)
            ax.streamplot(X, Y, U, V, color='white', density=1.0, linewidth=0.8)

            if (y_real == idx).sum() > 0:
                h_cls = h_real[y_real == idx]
                ax.scatter(h_cls[:100,0], h_cls[:100,1], c='cyan', s=10, alpha=0.8)

            digit = cfg.target_digits[idx]
            ax.set_title(f"Digit {digit} (Ep {frame_epoch})")
            ax.set_xlim(-1.8, 1.8); ax.set_ylim(-1.8, 1.8)

    try:
        ani = animation.FuncAnimation(fig, update, frames=sorted_epochs, interval=400)
        ani.save(os.path.join(save_dir, 'cognitive_evolution.gif'), writer='pillow', fps=2)
        print("‚úÖ GIF animation saved.")
    except Exception as e:
        print(f"‚ö†Ô∏è GIF generation failed (check ffmpeg/pillow): {e}")
    plt.close()

# --- ÂäüËÉΩ 3: Âõ†Êûú‰ª™Ë°®Áõò ---
def plot_causal_dashboard(logs, supervisor_log, save_dir):
    print("üìä Plotting causal dashboard...")
    fig = plt.figure(figsize=(14, 10))
    gs = gridspec.GridSpec(2, 1, height_ratios=[1, 1])

    # Upper: Accuracy & RPE
    ax1 = plt.subplot(gs[0])
    if len(logs['acc']) > 0:
        ax1.plot(logs['acc'], label='Accuracy', color='#1f77b4', linewidth=2)
        ax1.set_ylabel('Accuracy (%)', color='#1f77b4')
        ax1.tick_params(axis='y', labelcolor='#1f77b4')

        if len(logs['rpe']) > 0:
            ax2 = ax1.twinx()
            ax2.plot(logs['rpe'], label='RPE', color='#ff7f0e', linestyle='--', linewidth=2)
            ax2.set_ylabel('Avg RPE', color='#ff7f0e')
            ax2.tick_params(axis='y', labelcolor='#ff7f0e')
        ax1.set_title("System 1 Dynamics")

    # Lower: Stability & AI Intervention
    ax3 = plt.subplot(gs[1], sharex=ax1)
    if len(logs['eigenvalue']) > 0:
        # Align sparse eigenvalues to epochs
        step = cfg.consult_interval if hasattr(cfg, 'consult_interval') else 5
        x_eigs = [min(i * step, len(logs['acc'])-1) for i in range(len(logs['eigenvalue']))]

        ax3.plot(x_eigs, logs['eigenvalue'], color='#9467bd', marker='o', linewidth=2, label='|Œª|')
        ax3.axhline(1.0, color='red', linestyle=':', label='Stability Threshold')
        ax3.set_ylabel('|Œª| (Lower is Stabler)', color='#9467bd')
        ax3.set_xlabel('Epoch')
        ax3.legend()

    # Add AI Annotations
    if supervisor_log:
        for entry in supervisor_log:
            try:
                parts = entry.split(':')
                ep = int(parts[0].replace('Ep', ''))
                action = parts[1].split('|')[0].strip()
                if "DECAY" in action:
                    ax3.axvline(x=ep, color='green', linestyle='--', alpha=0.5)
                    ax3.text(ep, 1.5, ' AI:Decay', rotation=90, color='green')
                elif "STOP" in action:
                    ax3.axvline(x=ep, color='red', linestyle='-', alpha=0.8)
                    ax3.text(ep, 1.5, ' AI:STOP', rotation=90, color='red')
            except: pass

    plt.tight_layout()
    plt.savefig(os.path.join(save_dir, 'causal_dashboard.png'), dpi=100)
    plt.close()
    print("‚úÖ Dashboard saved.")

# --- ÂäüËÉΩ 4: ÁΩÆ‰ø°Â∫¶ÂàÜÂ∏É ---
def plot_behavioral_confidence(perception_net, cognitive_net, readout_net, loader, save_dir):
    print("üìä Analyzing confidence...")
    perception_net.eval(); cognitive_net.eval(); readout_net.eval()
    correct, incorrect = [], []

    try:
        with torch.no_grad():
            for x, y in loader:
                x = x.to(device)
                h = cognitive_net(perception_net(x), torch.zeros(x.size(0), cfg.hidden_size).to(device))
                probs, _ = readout_net(h)
                p_max, act = torch.max(probs, dim=1)
                mask = (act == y.to(device))
                correct.extend(p_max[mask].cpu().numpy())
                incorrect.extend(p_max[~mask].cpu().numpy())

        plt.figure(figsize=(8, 5))
        if len(correct) > 0:
            sns.histplot(correct, color='green', label='Correct', kde=True, stat='density', alpha=0.5, bins=20)
        if len(incorrect) > 0:
            sns.histplot(incorrect, color='red', label='Incorrect', kde=True, stat='density', alpha=0.5, bins=20)
        plt.title("Decision Confidence Distribution")
        plt.savefig(os.path.join(save_dir, 'confidence_dist.png'), dpi=100)
        plt.close()
        print("‚úÖ Confidence plot saved.")
    except Exception as e:
        print(f"‚ö†Ô∏è Confidence plot skipped: {e}")

# --- ‰∏ªÊâßË°åÈÄªËæë ---
if 'dynamics_history' in locals() and len(dynamics_history) > 0:
    save_static_snapshots(dynamics_history, perception_net, cognitive_net, readout_net, plot_dir)
    generate_evolution_gif(dynamics_history, perception_net, cognitive_net, readout_net, plot_dir)
    plot_causal_dashboard(logs, supervisor_log, plot_dir)
    plot_behavioral_confidence(perception_net, cognitive_net, readout_net, test_loader, plot_dir)
else:
    print("‚ö†Ô∏è Visualization skipped: No dynamics history found (Train loop didn't run?).")
```
